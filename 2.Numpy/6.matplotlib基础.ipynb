```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
cosy = np.cos(x)
plt.plot(x, y)
plt.plot(x, cosy, linestyle='--') #曲线颜色和形状
plt.xlim(-5, 15) #x轴的长度
plt.ylim(0, 1.5) #y轴的大小
plt.show()
```

```python
plt.plot(x, y, label = 'sin(x)')# lable表示这条线的名字，在下面加上plt.legend()就可以显示了
plt.plot(x, cosy, linestyle='--',label = 'cos(x)' ) #曲线颜色和形状
plt.legend()
plt.axis([-1, 11, -2, 2]) #直接调节x,y轴距离 （前两个为x的范围，后两个为y的范围）
plt.xlabel('x axis')
plt.ylabel('y value')#设置x,y轴的含义
plt.title("sin and cos")
plt.show()
```

## SCatter Plot

```python
plt.scatter(x, y) #散点图
plt.scatter(x, cosy)
plt.show()
```

```python
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)
plt.scatter(x, y, alpha = 0.5) #alpha设置透明程度
plt.show()
```
