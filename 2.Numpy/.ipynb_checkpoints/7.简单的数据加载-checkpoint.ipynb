```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris.keys()
```

```python
iris.feature_names
```

```python
X = iris.data[:, :2]
X.shape
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
y = iris.target #将不同品种进行不同颜色区分
plt.scatter(X[y ==0, 0], X[y == 0, 1], color = 'red', marker='o') # 点为红色，o型
plt.scatter(X[y ==1, 0], X[y == 1, 1], color = 'blue', marker='+')
plt.scatter(X[y ==2, 0], X[y == 2, 1], color = 'green', marker='x')
plt.show()
```

```python
X = iris.data[:, 2:]
y = iris.target #将不同品种进行不同颜色区分
plt.scatter(X[y ==0, 0], X[y == 0, 1], color = 'red', marker='o') # 点为红色，o型
plt.scatter(X[y ==1, 0], X[y == 1, 1], color = 'blue', marker='+')
plt.scatter(X[y ==2, 0], X[y == 2, 1], color = 'green', marker='x')
plt.show()
```
