### 梯度下降算法模拟

```python
import numpy as np 
import matplotlib.pyplot as plt

plot_x = np.linspace(-1, 6, 141)
plot_y = (plot_x -2.5) **2 -1

plt.plot(plot_x, plot_y)
plt.show()
```

```python
def dJ(theta): #即对plot_y进行求导
    return 2 * (theta -2.5)
```

```python
def J(theta):
    return (theta - 2.5 ) **2 - 1
```

```python
eta = 0.1 #步长
theta = 0.0 #初始点
epsilon = 1e-8 #前后两个损失函数的差值之间的间距
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta -eta * gradient
    if(abs(J(last_theta) - J(theta)) < epsilon):
        break

print(theta)
print(J(theta))
    
```

```python
#绘制下降曲线
theta = 0.0 #初始点
theta_history = [theta]
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta -eta * gradient
    theta_history.append(theta)
    if(abs(J(last_theta) - J(theta)) < epsilon):
        break

plt.plot(plot_x, J(plot_x))
plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'r', marker='+')
plt.show()
```

```python
#函数封装
def gradient_descent(inital_theta, eta, epsilon = 1e-8):
    theta = inital_theta
    theta_history.append(inital_theta)
    while True:
        gradient = dJ(theta)
        last_theta = theta
        theta = theta -eta * gradient
        theta_history.append(theta)
        if(abs(J(last_theta) - J(theta)) < epsilon):
            break

def plot_theta_historty():
    plt.plot(plot_x, J(plot_x))
    plt.plot(np.array(theta_history), J(np.array(theta_history)), color = 'r', marker='+')
    plt.show()
```

```python
eta = 0.01
theta_history = []
gradient_descent(0.0, eta)
plot_theta_historty()
```

```python
def gradient_descent(inital_theta, eta, n_iters = 1e4, epsilon = 1e-8): #n_iters设置循环的次数，不让循环无限的执行
    theta = inital_theta
    theta_history.append(inital_theta)
    i_iters = 0
    while i_iters < n_iters:
        gradient = dJ(theta)
        last_theta = theta
        theta = theta -eta * gradient
        theta_history.append(theta)
        if(abs(J(last_theta) - J(theta)) < epsilon):
            break
        i_iters += 1
```
