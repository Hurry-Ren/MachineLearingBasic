```python
import numpy as np
from sklearn import datasets

digits = datasets.load_digits()
X = digits.data
y = digits.target.copy()

y[digits.target == 9] = 1
y[digits.target != 9] = 0

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 666)

from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
decision_score = log_reg.decision_function(X_test)
```

```python
#从最小到最大逐渐移动，绘制精准率和召回率的变化曲线
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
import matplotlib.pyplot as plt

precisions = []
recalls = []
thresholds = np.arange(np.min(decision_score), np.max(decision_score), 0.1)
for threshold in thresholds:
    y_predict = np.array(decision_score >= threshold, dtype = 'int')
    precisions.append(precision_score(y_test, y_predict))
    recalls.append(recall_score(y_test, y_predict))
#绘制曲线    
plt.plot(thresholds, precisions)
plt.plot(thresholds, recalls)
plt.show() #红色为召回率
```

#### Precision-Recall曲线

```python
plt.plot(precisions, recalls)
plt.show() #随着精确率的上升，召回率在不断下降
#可以看到在某个点，召回率会明显下降很快，而在这个点之前，会有一些精确率和召回率相对来说比较平衡的点
```

### sklearn中的Precision-Recall曲线

```python
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(y_test, decision_score) #返回三个参数(精确率,召回率，步长)
precisions.shape
```

```python
thresholds.shape #因为最大的threshold对应的精确率为1、召回率为0是，没有对应的threshold，故少一个
```

```python
plt.plot(thresholds, precisions[:-1])
plt.plot(thresholds, recalls[:-1])
plt.show() #由于sklearn中包装的方法并没有从最小一直到最大这样取，故结果和我们不同
```
